Langkah selanjutnya adalah melakukan modeling terhadap data training yang telah kita buat. Setelah melalui serangkaian percobaan dengan menggunakan beberapa model Machine Learning, didapatkan bahwa algoritma Ada Boost Classifier memiliki accuracy score cukup tinggi yaitu 92%. Oleh karena itu, saya memilih menggunakan algoritma ML ini.

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB

# clf= SVC(kernel='linear')
# clf= KNeighborsClassifier(n_neighbors=19, weights='uniform')
# clf= MLPClassifier(alpha=1)
# clf= GradientBoostingClassifier()
# clf= RandomForestClassifier(n_estimators=200)
clf=AdaBoostClassifier(n_estimators=500)
result=cross_val_score(clf, df_train, y, scoring='roc_auc', cv=10)
result.mean()
```

```python
clf.fit(df_train,y) #melakukan fitting terhadap data train
```

```python
y_pred=clf.predict(df_test) #melakukan prediksi pada data test menggunakan model yang telah dibangun sebelumnya
y_pred
```
